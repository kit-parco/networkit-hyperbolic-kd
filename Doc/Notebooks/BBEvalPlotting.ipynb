```python
def simpleStat(c, title, xlabel, ylabel, query):
    c.execute(query)

    #Fetch data
    data = c.fetchall()
    x1 = list(map(lambda d: d[0], data))
    y1 = list(map(lambda d: d[1], data))

    #Set x axis text
    locs = tuple([x for x in range(len(data))])
    labels = tuple([data[x][0] for x in range(len(data))])
    plt.xticks(locs, labels)

    #Rotate x axis text
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=90)

    #Plot it
    plt.plot(range(len(data)), y1, 'ro-')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
def getPropertyQuery(graphFile, prop):
    return "SELECT algorithms.name AS algorithm, " + prop + " FROM graphs" + " JOIN properties ON graphs.id=properties.graphId" + " JOIN algorithms ON algorithms.id=properties.algorithmId" + " WHERE graphs.name='" + graphFile + "'"
```

```python
conn = sqlite3.connect('../../../ComplexNetworkBackbones/evaluation/output/backbones.db')
c = conn.cursor()
```

```python
graphs = ["./input/Caltech36.graphml", "./input/kitEmail.graphml"]
props = ["clusteringCoefficient", "degreeDistCoefficient", "cpvDistanceFromOriginalNormalized", "keptEdgesPercent", "graphStructuralRandMeasure", "runningTime"]
```

###### Simple stats

```python
for prop in props:
    for graph in graphs:
        title = graph + " - " + prop
        xlabel = "algorithm"
        ylabel = prop
        simpleStat(c, title, xlabel, ylabel, getPropertyQuery(graph, prop))
```

###### Degree distributions

```python
def compareDegreeDistributionsFromFileList(graphFiles):
    for graphFile in sorted(graphFiles):
        if "PROPERTIES" in graphFile:
            continue
        
        gLabel = graphFile.split("_")[1].replace(".txt", "")
        G = readGraph(graphFile, Format.METIS)
        dd = properties.degreeDistribution(G)
        plt.plot(dd, label=gLabel)

    plt.xscale("log")
    plt.xlabel("degree")
    plt.yscale("log")
    plt.ylabel("number of nodes")

    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    plt.show()
```

###### Degree distributions (selected files only)

```python
def compareDegreeDistributions(basePath, suffixes):
    graphFiles = sorted(list(map(lambda x: basePath.replace("SUFFIX", x), suffixes)))
    return compareDegreeDistributionsFromFileList(graphFiles)
```

Hypothesis: The distributions of the two samples are NOT the same.

If the K-S statistic is small or the p-value is high, then we cannot reject the hypothesis that the distributions of the two samples are the same. p is a probability and indicates how probable it is to get samples like the given ones given that the null hypothesis (the distributions are the same) is true.

## Generic function for convenient comparison of two distributions

```python
def compareDistributions(dist1, dist2):
    plt.plot(dist1, label='Original')
    plt.plot(dist2, label='Sparsified')
    plt.legend()
    plt.show()
    
    #Calculate KS
    ks, p = stats.ks_2samp(dist1, dist2)
    print("The hypothesis that the two distributions are the same cannot be rejected if the K-S statistic is small or if the p-value is high. \n")
    print("K-S statistic: ", ks)
    print("p-value: ", p)
```

## Distribution of Clustering Coefficients

Let's get the distribution C_d of the local clustering coefficient (per degree). C_d is defined as the average C_v (fraction of edges that exist) over all nodes of degree d.

```python
def getCCPerDegree(inputGraph):
    localCCs = properties.ClusteringCoefficient.exactLocal(inputGraph)
    maxDegree = max([inputGraph.degree(n) for n in inputGraph.nodes()])
    ccPerDegree = list(map(lambda d : np.average([localCCs[n] for n in inputGraph.nodes() if inputGraph.degree(n) == d]), range(0, maxDegree + 1)))
    #We map nan to 0. is this the way to go??
    ccPerDegree = list(map(lambda avg : 0.0 if math.isnan(avg) else avg, ccPerDegree))
    return ccPerDegree
```

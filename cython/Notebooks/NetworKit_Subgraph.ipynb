# Subgraph

This notebook illustrates the class Subgraph and how to use it to generate Subgraphs.

## Preparation

```python
cd ~/workspace/NetworKit/cython/
```

```python
from NetworKit import *
from graph import Subgraph
```

Subgraph generation will be demonstrated with the `PGPgiantcompo` graph. Feel free to try different graphs.

```python
G = readGraph("../input/PGPgiantcompo.graph")
```

To start, we use the `community` module to get the communities from graph `G`.

```python
communities = community.detectCommunities(G)
```

Now we take a look at the sizes of the different clusters.

```python
communities.clusterSizeMap()
```

We decide that community `2` is of interest, so we store the node IDs of that community in `c2`. Afterwards we use the class `Subgraph` to generate the graph `g2` induced by the nodes of cluster `2`.

```python
c2 = communities.getMembers(2)
sg = Subgraph()
g2 = sg.fromNodes(G,c2)
```

As we can see, the amount of nodes of new graph match with those of `c2`. Furthermore, graph `g2` contains every intra-cluster edge of community `c2`.

```python
g2.numberOfNodes()
```

We can now use the subgraph `g2` for further analysis, e. g. drawing:

```python
viztasks.drawGraph(g2)
```

or further community detection:

```python
communities2 = community.detectCommunities(g2)
```

```python
viztasks.drawCommunityGraph(g2,communities2)
```
